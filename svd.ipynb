```python
from sklearn.decomposition import TruncatedSVD
```

```python
for file in os.listdir('bucket/anndata'):
  adata = anndata.read_h5ad('bucket/anndata/'+file)
  connectivities = adata.obsp['connectivities']
  svd = TruncatedSVD(n_components=32)
  svd_result = svd.fit_transform(connectivities)
  output_dir = "bucket/svd/"

  np.savetxt(output_dir+'svd_'+file.strip("h5ad")+'csv', svd_result, delimiter=',')

```
